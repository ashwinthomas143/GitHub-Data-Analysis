The problem statement involves creating a solution that deals with customer data, compound interest calculation, and identifying customers whose savings after interest exceed a certain threshold. Here's a brief background for each step:

1. Creating Customer Dictionary:
   - A dictionary is created to store customer information.
   - The dictionary consists of 10 customers, each identified by a unique 7-digit customer ID.
   - For each customer, the dictionary stores the following values: first name, last name, account type (savings or checking), and the amount.

2. Calculating Compound Interest:
   - Three new dictionaries are created to store compound interest values for each customer ID.
   - The customer ID serves as the key, while the values represent the compound interest over different time periods (1 year, 3 years, and 5 years).
   - The interest rate used for the calculations is 17%, compounded annually.

3. Identifying Customers with Exceeding Savings:
   - From the customer dictionary, the program identifies customer IDs whose 5-year savings after interest are 1.5 times or more their original savings.
   - The program compares the final savings after 5 years (including interest) with the original savings.
   - If the final savings exceed or equal 1.5 times the original savings, the customer ID is listed.

Overall, the goal of this solution is to manage customer data, calculate compound interest, and identify customers who have achieved significant savings growth over time.

### 1.Create a Dictionary of 10 customer with unique 7digit customer ID as the key of dictionaries. Then store the following values as dictionaries - first name, last name, account type (savings or chequing (5 each type)) and amount as values for each customer id key.

```python
bank_dict = {
	"0000001":{"firstName":"Daniel","lastName":"Brown","accountType":"Savings"},
	"0000002":{"firstName":"Olivia","lastName":"Johnson","accountType":"Checking"},
	"0000003":{"firstName":"Emma","lastName":"Smith","accountType":"Savings"},
	"0000004":{"firstName":"William","lastName":"Miller","accountType":"Checking"},
	"0000005":{"firstName":"Sophia","lastName":"Davis","accountType":"Savings"},
	"0000006":{"firstName":"David","lastName":"Anderson","accountType":"Savings"},
	"0000007":{"firstName":"Michael","lastName":"Wilson","accountType":"Checking"}
}

for key in bank_dict.keys():
    investment_amount = float(input(f"Please enter how much {bank_dict[key]['firstName']} wants to invest :\n"))
    bank_dict[key]["amount"] = str(round(investment_amount,2))

print("The following is the customer information for the first 7 customers, banking at TD Bank:\n")

for k,v in bank_dict.items():
    print(f"Customer ID : {k}")
    print(f"First Name: {v['firstName']}")
    print(f"Last Name: {v['lastName']}")
    print(f"Account Type: {v['accountType']}")
    print(f"Amount: {v['amount']}")
    print()
```

### 2.Create 3 new dictionary saving the customer ID as keys and 1,3, and 5-year compound interest as values as 3 dictionaries.

```python
initial_investment = {}

one_year_interest = {}
one_year_return = {}

three_year_interest = {}
three_year_return = {}

five_year_interest = {}
five_year_return = {}


interest_rate = float(input("Please enter an interest rate."))

for k,v in bank_dict.items():
    initial_investment[k] = round(float(v["amount"]))
    
    one_y_interest = float(v["amount"])*(1+interest_rate/100) ** 1
    one_year_interest[k] = round(one_y_interest) - round(float(v["amount"]))
    one_year_return[k] = round(one_y_interest)

    three_y_interest = float(v["amount"])*(1+interest_rate/100) ** 3
    three_year_interest[k] = round(three_y_interest) - round(float(v["amount"]))
    three_year_return[k] = round(three_y_interest)
    
    five_y_interest = float(v["amount"])*(1+interest_rate/100) ** 5
    five_year_interest[k] = round(five_y_interest) - round(float(v["amount"]))
    five_year_return[k] = round(five_y_interest)
    
print(f"Our current customer balances :\nInitially")
print(initial_investment)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal interest:\nafter 1 year")
print(one_year_interest)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal returns:\nafter 1 year")
print(one_year_return)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal interest:\nafter 3 years")
print(three_year_interest)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal returns:\nafter 3 years")
print(three_year_return)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal interest:\nafter 5 years")
print(five_year_interest)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal returns:\nafter 5 years")
print(five_year_return)
print()
```

### 3.List down customer IDs whose 5-year savings after interest is 1.5 times or more their original savings.Interest is 17% compounded annually.

-	Give your output as a HTML page having logo of the bank of your choice.
-	Give a brief background of the problem statement.
-	Beautify your output html to be client ready and presentable.

```python
initial_investment = {}

five_year_interest = {}
five_year_return = {}

interest_rate = 17

for k,v in bank_dict.items():
    initial_investment[k] = round(float(v["amount"]))
    five_y_interest = float(v["amount"])*(1+interest_rate/100) ** 5
    five_year_interest[k] = round(five_y_interest) - round(float(v["amount"]))
    five_year_return[k] = round(five_y_interest)

print(f"Our current customer balances :\nInitially")
print(initial_investment)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal interest:\nafter 5 years")
print(five_year_interest)
print()

print(f"If a customer invests at {interest_rate} % with our bank, the following would be the ideal returns:\nafter 5 years")
print(five_year_return)
print()

print(f"The following is a list of customers that earned 1.5 times their initial investment, after investing with TD Bank:")
for k,v in initial_investment.items():
    initial_amount = v
    final_return = five_year_return[k]
    
    if final_return / initial_amount >= 1.5:
        print(f"Customer ID: {k}")
        print(f"Initial Investment: {initial_amount}")
        print(f"Final Return on Investment: {final_return}")
        print()
        
```
